#Tipe Nilai Atribut


1.   Nominal : digunakan sebagai nama dari sesuatu bentuk data. Jenis tipe atribut ini hanya bisa dilakukan operasi distinctness (membandingkan sama atau tidaknya suatu data). Contoh : nama orang, alamat, nomor ktp, kode pos, dan warna mata dll. Operasi yang bisa dilakukan adalah distinctness.
2.   Ordinal : berupa rangking/tingkatan yang bisa diperbandingkan tingkatannya. Contoh : grade tinggi, sedang, rendah. Operasi yang bisa dilakukan adalah distinctness & order.
3.  Interval : memiliki nilai yang sudah pasti (tetap) diantara tiap data interval. Contoh : tanggal kalender (diantara dua tanggal tidak ada tanggal lain), temperatur dalam Celcius atau Fahrenheit. Operasi yang bisa dilakukan adalah distinctness, order & addition.
4. Ratio : Memiliki nilai 0 (nol). Contoh : temperatur dalam Kelvin, panjang, waktu, massa dan jumlah. Operasi yang bisa dilakukan adalah semua operasi yang ada.
